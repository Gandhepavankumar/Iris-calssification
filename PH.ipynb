```python
import pandas as pd
from scipy.stats import chi2_contingency

# Assuming input_variables is a list of column names you want to analyze
input_variables=['srcipt', 'sport', 'dstip', 'dsport', 'proto',
                 'state', 'dur', 'sbytes', 'dbytes', 'sttl', 'dttl', 'sloss', 'dloss', 'service', 'Sload', 'Dload',
                 'Spkts', 'Dpkts', 'swin', 'dwin', 'stcpb', 'dtcpb', 'smeansz', 'dmeanz', 'trans_depth', 'res_bdy_len', 'Shoot',
                 'Djit', 'Estimates', 'Ltime', 'Sintpkt', 'Dintpkt', 'tcprtt', 'synack', 'thank you', 'is_sm_ips_ports', 'ct_state_ttl',
                 'ct_flw_http_mthd', 'is_ftp_login', 'ct_ftp_cmd', 'ct_srv_src', 'ct_srv_dst', 'ct_dst_ltm', 'ct_src_ ltm', 'ct_src_dport_ltm',
                 'ct_dst_sport_ltm', 'ct_dst_src_ltm']  # Your list of input variables
new_list = []

# Loop through each input variable
for input_var in input_variables:
    contingency_table = pd.crosstab(data_train[input_var], data_train['Label'])

    # Check if the contingency table is empty or not
    if contingency_table.empty:
        print(f"Contingency table for {input_var} is empty.")
        continue  # Skip further processing for this input variable

    stat, p, dof, expected = chi2_contingency(contingency_table)

    alpha = 0.05  # Significance level
    if p <= alpha:
        print(f"{input_var} and 'Label' are dependent.")
    else:
        print(f"{input_var} and 'Label' are independent.")
        new_list.append(input_var)

print("Variables with independent relationship:", new_list)
```
